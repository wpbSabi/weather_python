```python
df = pd.read_csv('../../data/temperature/temperature_97218.csv')
df = df.assign(year = pd.to_datetime(df['DATE']).dt.year)
df = df.assign(month = pd.to_datetime(df['DATE']).dt.month)
df = df.assign(day = pd.to_datetime(df['DATE']).dt.dayofyear)
df1 = df[df['year']<2025]
assert len(df['STATION'].unique()) == 1
```
